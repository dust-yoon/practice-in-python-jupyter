## (2) 논리연산

* and 연산 즉, & 의 경우 모든 조건이 참(True)이어야 결과가 True로 나옴
* 어느 하나라도 조건이 거짓(False)이면 결과는 False
* or 연산 즉, | 의 경우 모든 조건이 거짓(False)이어야 결과가 False로 나옴
* 어느 하나라도 조건이 참(True)이면 결과는 True

#### 비교연산과 논리연산
###### 결과값이 모두 bool값 형태. 즉 True와 False로 나옴

```python
(7<=2) & (15>5) # (조건1) 거짓. (조건2) 참 = False
```

```python
(7<=2) | (15>5) # (조건1) 거짓. (조건2) 참 = True
```

#### ⅰ) 부분합 기본 코드
* 변수명.groupby(by='그룹').데이터형식(예시.sum())
* 변수 데이터 모두를 그룹으로 요약. 이때 요약 정보는 '데이터형식' 대로 나타낸다.
* 데이터 형식: sum(합계), mean(중앙값), count(수량), min(최솟값), max(최댓값), var(분산), std(표준편차)

```python
nana
```

```python
nana.groupby(by='부서').sum( )
# nana 데이터 모두를 '부서' 그룹으로 요약하되 요약 정보는 합계로 나타내라
```

```python
nana.groupby(by='부서')[['봉급']].mean( ) # 요약 정보는 평균값
```

```python
nana.groupby(by='부서')[['봉급']].count( ) # 요약 정보는 개수
```

```python
nana.groupby(by='부서')[['봉급']].max( ) # 요약 정보는 최댓값
```

```python
nana.groupby(by='부서')[['봉급']].min( ) # 요약 정보는 최솟값
```

```python
nana.groupby(by='부서')[['봉급']].var( ) # 요약 정보는 분산
```

```python
nana.groupby(by='부서')[['봉급']].std( ) # 요약 정보는 표준편차
```

#### ⅱ) 부분합 데이터 활용
* 변수명[['열1', '열2']].groupby(by='그룹').sum( )
