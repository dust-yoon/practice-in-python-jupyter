## 데이터 정렬하기
#### 변수명.sort_values("기준 삼을 열의 이름")
* nana.sort_values('부서', ascending=False) '부서'열의 값을 내림차순 기준으로 데이터들을 정렬 시킴
#### 변수명.sort_index() 인덱스를 기준으로 데이터 정렬
* nana.sort_index( ascending=False ) 인덱스를 내림차순 기준으로 데이터들을 정렬 시킴
* ascending = True면 오름차순 기

```python
nana.sort_values('부서') # '부서'의 값 기준으로 데이터들을 정렬 시킴
```

```python
nana.sort_values('부서', ascending=False)
```

### 부분합 구하기

```python
nana
```

```python
nana.groupby(by='부서').sum( )
# nana 데이터 모두를 '부서' 그룹으로 요약하되 요약 정보는 합계로 나타내라
```

```python
# nana[['봉급'], nana.groupby(by='부서').sum( )
# SyntaxError: incomplete input
nana[['봉급', '부서']].groupby(by='부서').sum( )
# nana의 열 중에서 '봉급'열과 '부서'열을 뽑아서 이것을 '부서' 그룹으로 합계 나타내라
```

```python
nana.groupby(by='부서')['봉급'].sum( ) # 숫자 칼럼들 중 봉급 관련 자료만 sum 시켜다오
# 둥근괄호: 함수 성격인 애들
# 꺽쇠괄호: 슬라이트 특정 부위를 드러내는 애들
```

```python
nana.groupby(by='부서')[['봉급']].sum( )
# 2차원 데이터(데이터프레임 만드는 법)
```

```python
pd.DataFrame( nana.groupby(by='부서')['봉급'].sum( ) )
# 겹괄호 허용안되는 코드들이 데이터 프레임을 이용하는 방법
```

```python
nana.groupby(by='부서')[['봉급']].mean( ) # 요약 정보는 평균값
```

```python
round( nana.groupby(by='부서')[['봉급']].mean( ),2 )
```

```python
aa = nana.groupby(by='부서')[['봉급']].mean( )
```

```python
round( aa, 2) # aa의 값을 반올림 해서 소수점 둘째 자리까지 나타내기
```

```python
round(nana.groupby(by='부서')[['봉급']].mean( ),2)
# 변수로 처리하지 않고 그냥 넣어도 되지만 오류 확인이 어렵고 가독성이 떨어짐
```

```python
nana.groupby(by='부서')[['봉급']].count( ) # 요약 정보는 개수
```

```python
nana.groupby(by='부서')[['봉급']].max( ) # 요약 정보는 최댓값
```

```python
nana.groupby(by='부서')[['봉급']].min( ) # 요약 정보는 최솟값
```

```python
nana.groupby(by='부서')[['봉급']].var( ) # 요약 정보는 분산
```

```python
nana.groupby(by='부서')[['봉급']].std( ) # 요약 정보는 표준편차
```

# 피벗 테이블 만들기
### 변수명.pivot_table(index=행기준 , columns=열기준 , values=데이터 영역 , aggfunc='mean','sum' ... )
* NaN Not a Number = 안에 값이 없다.

```python
nana
```

```python
nana.pivot_table(index='직책' , columns='부서' , values='봉급' , aggfunc='count' )
```

```python
nana_pt = nana.pivot_table(index='직책' , columns='부서' , values='봉급' , aggfunc='sum' )
nana_pt
```

```python
nana.pivot_table(index=['지역', '직책'] , columns='부서' , values='봉급' , aggfunc='sum' )
```

```python
nana_pt.sum()
```

```python
nana_pt.sum(axis=0) # (가로기준)세로 방향으로 값을 하나씩 더 내줌/index 값
```

```python
nana_pt.sum(axis=1) # (세로기)가로 방향으로 값을 하나씩 더 내줌/columns
```

```python
pd.DataFrame(nana_pt.sum(axis=1))
```
