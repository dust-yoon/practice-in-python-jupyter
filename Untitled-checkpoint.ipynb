# 데이터 탐색용 코드들
### 데이터 세트를 이용해서 다른 작업(후속작업)을 하고 싶다면 반.드.시. 변수에 담을 것
##### 변수명 = 어떤 코드
* 예) gaga = pd.read_csv('se_sales.txt', sep='\t', skiprows=7)

##### 미리보기 용 코드
* 변수명.head( ) 미리보기 앞부분 기본 5줄 보여 줌
* 변수명.tail( ) 미리보기 뒷부분 기본 5줄 보여 줌
* 변수명.head(3), 변수명.head(n=2), tail(2) 등으로 보고 싶은 줄 갯수 지정 가능

#### 데이터 정보를 보여주는 코드
##### 변수명.info( )
* 해당 변수의 전체 틀 유형(예: 데이터프레임), 컬럼 이름과 갯수, 행 갯수, 각 컬럼별 들어간 데이터(원소)의 유형(예: 정수형, 실수형, 문자형..), 빈 값 유무..

##### 변수명.describe( )
* 해당 변수의 데이터 분포 정보 제공, 열의 값(즉, 원소)이 숫자형태인 경우, 값의 갯수, 평균값, 표준편차, 최솟값, 최댓값, 25%, 50%, 75%에 해당하는 값

##### 변수명.describe( include = 'all')
* 열의 값이 숫자로 되어 있던 문자로 되어 있던 해당 값의 분포 정보를 제공, 열의 값이 문자인 경우, unique(값이 몇개의 종류인지), top(가장 빈도 높은 데이터의 첫 번째 값), freq(가장 빈도가 높은 데이터 값)
