### 행 단위로 데이터 추출
##### 순번: 데이터 행의 순서, 정수만으로 조회나 추출됨, DataFrame에서 확인할 수 없음(컴퓨터가  지 혼자 속으로 세고 있음)
* 순번은 행의 순번, 열의 순번이 존재
* 사용자가 임의로 순번 바꿀 수 없음
* 순번은 항상 0부터 시작함

##### 인덱스: DataFrame에서 눈에 보임, 행과 행을 구별하는 용도로 많이 사용, 인덱스는 사용자가 정수, 문자(열) 등으로 줄 수 있음

##### loc
* loc 는 인덱스를 기준으로 행 데이터를 추출
* 변수명.loc [ 인덱스 ] # 행 1개만 추출할 때
* 변수명.loc [ [인덱스a, 인덱스b, 인덱스c] ]  # 행 여러개 추출할 때
* 변수명.loc [ 인덱스a : 인덱스c ]  # 연이은 행 여러개 추출할 때, 인덱스a 부터 인덱스c 까지의 모든 행 나옴
* 인덱스가 숫자가 아니고 문자일때는 항상 따옴표 써야 함 # 예) 변수명.loc["인덱스"]
* nana.loc [ 0: 15: 5 ] # 인덱스 0부터 15까지 5개씩 건너 띄어라. 결과는 인덱스 0, 5, 10, 15에 해당하는 4개의 행 나옴
* 

  ##### iloc
* iloc 는 순번을 기준으로 행 데이터를 추출
* 변수명.iloc [ 순번 ] # 행 1개만 추출할 때
* 변수명.iloc [ [순번0, 순번3, 순번21] ]  # 행 여러개 추출할 때
* 변수명.iloc [ 순번3 : 순번5 ]  # 연이은 행 여러개 추출할 때, 순번3 부터 순번5 "직전"까지의 모든 행 나옴, 즉 순번3, 순번4에 해당하는 행만 나옴
* nana.iloc[ 0: 15 : 5 ]  # 순번 0부터 15 직전까지 5개씩 건너 띄어라. 결과는 순번 0, 5, 10에 해당하는 3개의 행 나옴

### 특정 행과 열 데이터 추출(슬라이싱)
* 변수명.loc [ 행정보 , 열정보]
* 변수명.iloc[ 행정보 , 열정보]

##### 변수명.loc [ 행인덱스, 열이름 ]
* 변수명.loc [ : , : ] 행 몽땅, 열 몽땅
* 변수명.loc [ [인덱스a, 인덱스b, 인덱스c] , [열이름a, 열이름p] ]
* 변수명[열이름] . loc[인덱스]
* 열이름과 인덱스가 숫자가 아니고 문자일때는 항상 따옴표 써야 함 # 예) 변수명.loc["인덱스a", ["열이름a", "열이름p"]]

##### 변수명.iloc [ 행순번, 열순번 ]
* 변수명.iloc [ : , : ] 행 몽땅, 열 몽땅
* 변수명.iloc [ [행순번0, 행순번3, 행순번21] , [열순번0, 열순번2] ]
* 변수명[열이름] . iloc[행순번]
* 열이름이 문자일때는 항상 따옴표 써야 함 # 예) 변수명["열이름"].iloc[행순번]
